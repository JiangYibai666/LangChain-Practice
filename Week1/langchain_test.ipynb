# Using LangChain as a thin-package call model for parameterized translation.

````python
from dotenv import load_dotenv, find_dotenv
# from langchain_anthropic import ChatAnthropic
from langchain_google_genai import ChatGoogleGenerativeAI
import os
print(os.getcwd())
````

````python
load_dotenv (find_dotenv ()) # read local .env file

# if not os.getenv("ANTHROPIC_API_KEY"):
#     raise ValueError("Cannot find ANTHROPIC_API_KEY，Check .env")

if not os.getenv("GEMINI_API_KEY"):
    raise ValueError("Cannot find GEMINI_API_KEY，Check .env")
````

````python
chat = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)
chat
````

````python
#create a prompt template
template_string = """Translate the text \
that is delimited by triple backticks \
into a style that is {style}. \
Keep the same paragraph breaks and line breaks as the source text. \
text: ```{text}```"""
````

````python
from langchain_core.prompts import ChatPromptTemplate
prompt_template = ChatPromptTemplate.from_template(template_string)
````

````python
target_style = """
Academic English \
in a logical and organized manner, using appropriate vocabulary and grammar to convey complex ideas effectively.
"""

text = """
In recent years, artificial intelligence systems have been widely adopted across various industries,
but many real-world deployments still face challenges such as data quality, model robustness, and interpretability.
These challenges highlight the gap between theoretical model performance and practical application constraints.
"""
````

````python
messages = prompt_template.format_messages (style=target_style, text=text)
response = chat.invoke(messages)
print(response.content)
````
