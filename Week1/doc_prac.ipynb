### Document Processing (Vector Database)

```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())  # read local .env file
if not os.getenv("GEMINI_API_KEY"):
    raise ValueError("GEMINI_API_KEY was not found. Check .env")
```

```python
from langchain_classic.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_community.document_loaders import CSVLoader
from langchain_community.vectorstores import DocArrayInMemorySearch
from IPython.display import display, Markdown
```

```python
# Returns a list of LangChain Document objects, suitable for vector stores and RAG retrieval; do not use pd.read_csv directly
file = 'products.csv'
loader = CSVLoader(file_path=file)
```

#### Turn the data into a semantically searchable knowledge base

```python
from langchain_classic.indexes import VectorstoreIndexCreator
```

```python
# Read the text, convert it to embeddings, and build the index
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
index = VectorstoreIndexCreator(
    embedding=embeddings,
    vectorstore_cls=DocArrayInMemorySearch,
).from_loaders([loader])
```

#### Use the model to read the documents as requested

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

query = "List all the products with smart home " \
"in a table in markdown and summarize each product in one sentence."
response = index.query(query, llm=llm)
display(Markdown(response))
```
