### Document Processing (Vector Database)

```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())  # read local .env file
if not os.getenv("ANTHROPIC_API_KEY"):
    raise ValueError("ANTHROPIC_API_KEY was not found. Check .env")
```

```python
from langchain_classic.chains import RetrievalQA
from langchain_anthropic import ChatAnthropic
from langchain_community.embeddings import FakeEmbeddings
from langchain_community.document_loaders import CSVLoader
from langchain_core.vectorstores import InMemoryVectorStore
from IPython.display import display, Markdown
```

```python
# Returns a list of LangChain Document objects, suitable for vector stores and RAG retrieval; do not use pd.read_csv directly
file = 'products.csv'
loader = CSVLoader(file_path=file)
```

#### Turn the data into a semantically searchable knowledge base

```python
from langchain_classic.indexes import VectorstoreIndexCreator
```

```python
# Build the index without external embedding APIs
embeddings = FakeEmbeddings(size=768)
index = VectorstoreIndexCreator(
    embedding=embeddings,
    vectorstore_cls=InMemoryVectorStore,
).from_loaders([loader])
```

#### Use the model to read the documents as requested

```python
import os
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv())

api_key = os.getenv("ANTHROPIC_API_KEY")
if not api_key:
    raise ValueError("ANTHROPIC_API_KEY was not found. Check .env")

llm = ChatAnthropic(
    model="claude-3-haiku-20240307",
    temperature=0,
    anthropic_api_key=api_key,
 )

query = "List all the products with Groceries " \
"in a table in markdown and summarize each product in one sentence."
response = index.query(query, llm=llm)
display(Markdown(response))
```
